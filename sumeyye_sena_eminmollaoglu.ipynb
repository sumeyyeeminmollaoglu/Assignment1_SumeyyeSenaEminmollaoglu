```python
import thinkplot
import thinkstats2
```

```python
ages=resp.fmarno
hist = thinkstats2.Hist(ages, label='Number of Marriage Done by Respondents')
thinkplot.Hist(hist)
thinkplot.Config(xlabel='Number of Marriage', ylabel='Number of Respondents')
```

```python
neverMarried = resp[resp.fmarno == 0]
married = resp[resp.fmarno > 0]

nevermarried_hist = thinkstats2.Hist(neverMarried.totincr, label='neverMarried')
married_hist = thinkstats2.Hist(married.totincr, label='married')
```

```python
width = 0.5
thinkplot.PrePlot(2)
thinkplot.Hist(nevermarried_hist, align='left', width=width)
thinkplot.Hist(married_hist, align='right', width=width)
thinkplot.Config(xlabel='Total Income', ylabel='Number of People')
```
